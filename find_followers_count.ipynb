# Problem Statement
Write a solution that will, for each user, return the number of followers.

Return the result table ordered by user_id in ascending order.

**Example 1:**

**Input:** 
Followers table:
| user_id | follower_id |
|---------|-------------|
| 0       | 1           |
| 1       | 0           |
| 2       | 0           |
| 2       | 1           |

**Output:** 
| user_id | followers_count |
|---------|-----------------|
| 0       | 1               |
| 1       | 1               |
| 2       | 2               |

# Intuition
The goal here is to count how many followers each user has. We'll use grouping to aggregate followers for each user and then count them.

# Approach

**Solution 1: Using Direct Aggregation**
1. **Group by User:** Group the data by `user_id` to count followers for each user.
2. **Count Followers:** Use `COUNT(follower_id)` to get the number of followers.
3. **Order Results:** Order the results by `user_id` in ascending order.

# Complexity

- **Time complexity:** 
  - O(n log n), where `n` is the number of rows in the `Followers` table. The grouping is linear, but sorting adds a logarithmic factor.

- **Space complexity:** 
  - O(n), where `n` is the number of rows in the `Followers` table. We need space to store the grouped and sorted results.

# Code

## Solution 1: Direct Aggregation
```sql
SELECT
    user_id,
    COUNT(follower_id) AS followers_count
FROM
    Followers
GROUP BY
    user_id
ORDER BY
    user_id ASC;
