# LeetCode Problem: Query Quality and Poor Query Percentage

## Description
We define query quality as:

- **The average of the ratio between query rating and its position.**

We also define poor query percentage as:

- **The percentage of all queries with rating less than 3.**

Write a solution to find each query_name, the quality, and poor_query_percentage.

Both quality and poor_query_percentage should be rounded to 2 decimal places.

Return the result table in any order.
## Example

**Input:** 
```plaintext
Queries table:
+------------+-------------------+----------+--------+
| query_name | result            | position | rating |
+------------+-------------------+----------+--------+
| Dog        | Golden Retriever  | 1        | 5      |
| Dog        | German Shepherd   | 2        | 5      |
| Dog        | Mule              | 200      | 1      |
| Cat        | Shirazi           | 5        | 2      |
| Cat        | Siamese           | 3        | 3      |
| Cat        | Sphynx            | 7        | 4      |
+------------+-------------------+----------+--------+

output table:
+------------+---------+-----------------------+
| query_name | quality | poor_query_percentage |
+------------+---------+-----------------------+
| Dog        | 2.50    | 33.33                 |
| Cat        | 0.66    | 33.33                 |
+------------+---------+-----------------------+



# Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The problem asks for two metrics based on query results: **quality** and **poor query percentage**. Quality is about how ratings correlate with result positions, while poor query percentage counts queries with low ratings. My initial approach is to aggregate data per query.

# Approach
<!-- Describe your approach to solving the problem. -->
**Quality**: We'll calculate the average ratio of rating to position for each query name. Lower positions are better, so this gives an idea of how well the search engine is performing for each query.

**Poor Query Percentage**: Determine how many results for each query have a rating less than 3, then express this as a percentage.

Use SQL to aggregate this data using `GROUP BY`, `AVG`, and `ROUND` functions.

# Complexity
**Time complexity:**  
O(n log m) where *n* is the number of rows and *m* is the number of unique query names due to sorting in `GROUP BY`.

**Space complexity:**  
O(m) for holding the grouped results, where *m* is the number of unique query names.

# Code
```sql
SELECT
  query_name,
  ROUND(AVG(rating / position), 2) AS quality,
  ROUND(AVG(rating < 3) * 100, 2) AS poor_query_percentage
FROM
  Queries
GROUP BY
  query_name;
