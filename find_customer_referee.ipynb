# SQL Query for Customers Not Referred by Customer with ID = 2

## Table Schema

- **Customer**
  - **id**: INT PRIMARY KEY
  - **name**: VARCHAR(255)
  - **referee_id**: INT

## Problem Statement

Find the names of customers that are not referred by the customer with `id = 2`. 
The result table can be in any order.

## Example Tables

Here's an example of how the `Customer` table might look:

```sql
+----+------+------------+
| id | name | referee_id |
+----+------+------------+
| 1  | Will | null       |
| 2  | Jane | null       |
| 3  | Alex | 2          |
| 4  | Bill | null       |
| 5  | Zack | 1          |
| 6  | Mark | 2          |
+----+------+------------+
Intuition
We need to identify customers whose referee_id is either NULL or not equal to 2. This means we're looking for customers who either have no referee or have a referee other than customer 2.

Approach
Select: We'll select the name column from the Customer table.
Filter: Use a WHERE clause to exclude customers with referee_id = 2.

Steps:
Select: Only the name of the customer.
Where: Apply the condition where referee_id is NULL or not equal to 2.

Complexity
Time Complexity: O(n), where n is the number of rows in the Customer table. Each row must be checked.
Space Complexity: O(1), as we're selecting one column without additional storage relative to input size.

SQL Query
Here's the SQL query to solve this problem:

SELECT name
FROM Customer
WHERE referee_id IS NULL OR referee_id != 2;

+------+
| name |
+------+
| Will |
| Jane |
| Bill |
| Zack |
+------+
